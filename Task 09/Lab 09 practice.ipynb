### Data Preprocessing


```python
df.info()
```

```python
df.drop('date', axis=1)
##yaha pr data hamara save ni horha isky liya hum 2 method krygy jo neechay hy
# 1. df = df.drop('date', axis=1)
df.drop('date', axis=1, inplace=True)
```

```python
df.drop('country', axis=1, inplace=True)
```

```python
#Mean , Mode , Median
## mean and medisan work only on numerical data
df['city'].mode()
## yaha 2 values arhi hy usky solution ky liya hum krygy neechy wala
```

```python
type(df['city'].mode())
```

```python
df['city'].mode()[0]
```

```python
df['city'].fillna(df['city'].mode()[0], inplace=True)
## agar bagir print ky hamary pass output arha hy to iska mtlb hy ky hamari memory my save ni horha hy
```

```python
def FILLNAOBJECTMODE(cols):
    for i in cols:
        df[i]= df[i].fillna(df[i].mode()[0])

columns = ['city', 'street', 'statezip']
FILLNAOBJECTMODE(columns)
```

```python
def FILLNAFLOATMODE(cols):
    for i in cols:
        df[i]= df[i].fillna(df[i].mean())


columns = ['price', 'condition', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement']
FILLNAFLOATMODE(columns)
```

```python
def FILLNAINTMODE(cols):
    for i in cols:
        df[i]= df[i].fillna(df[i].mode()[0])


columns = ['bedrooms', 'bathrooms', 'floors', 'waterfront', 'view', 'yr_built', 'yr_renovated']
FILLNAINTMODE(columns)
```
