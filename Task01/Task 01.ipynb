## Question: Hangman Game in Python 

Write a Python program to create a **Hangman Game** dynamically .

```python
import random

HANGMAN_PICS = [
    '''
    +---+
      O   |
     /|\\  |
     / \\  |
         ===''','''
    +---+
      O   |
     /|\\  |
     /    |
         ===''', '''
    +---+
      O   |
     /|\\  |
          |
         ===''', '''
    +---+
      O   |
     /|   |
          |
         ===''', '''
    +---+
      O   |
      |   |
          |
         ===''', '''
    +---+
      O   |
          |
          |
         ===''', '''
    +---+
          |
          |
          |
         ==='''
]

word_list = {
    "Animals": ["tiger", "lion", "elephant", "giraffe", "monkey"],
    "Fruits": ["apple", "banana", "mango", "orange", "grapes"]
}

lives = 6
category = random.choice(list(word_list.keys()))   
chosen_word = random.choice(word_list[category])   
display = ['_'] * len(chosen_word)  

print("Welcome to Hangman!")
print(f"Category: {category}")

game_over = False
guessed_letters = []  

while not game_over:
    print(f"Lives left: {lives}")
    print(' '.join(display))  

   
    guessed_letter = input("Guess a letter: ").lower()

    if guessed_letter in guessed_letters:
        print(f"You already guessed the letter '{guessed_letter}'. Try again.")
        continue

    guessed_letters.append(guessed_letter)

    if guessed_letter in chosen_word:
        for position in range(len(chosen_word)):
            if chosen_word[position] == guessed_letter:
                display[position] = guessed_letter
    else:
        lives -= 1
        print(f"'{guessed_letter}' is not in the word.")
    
    print(HANGMAN_PICS[lives])

   
    if '_' not in display:
        game_over = True
        print("Congratulations! You won!")
    elif lives == 0:
        game_over = True
        print("You lose! The word was:", chosen_word)
```

## Question: To-Do List Application in Python

Write a Python program to create a simple **To-Do List Application** dynamically that allows the user to manage tasks.  

```python
tasks = []


def addTask():
    task = input("Please enter a task: ")
    tasks.append(task)
    print(f"Task '{task}' added to the list.")


def listTasks():
    if not tasks:
        print("There are no tasks currently.")
    else:
        print("Current Tasks:")
        for index, task in enumerate(tasks, start=1):
            print(f"Task #{index}. {task}")


def deleteTask():
    listTasks()
    if not tasks:
        return
    try:
        taskToDelete = int(input("Enter the # to delete: "))
        index = taskToDelete - 1
        if 0 <= index < len(tasks):
            removed = tasks.pop(index)
            print(f"Task {taskToDelete} ('{removed}') has been removed.")
        else:
            print(f"Task #{taskToDelete} was not found.")
    except ValueError:
        print("Invalid input. Please enter a number.")


if __name__ == "__main__":

    print("Welcome to the to do list app :)")
    while True:
        print("\n")
        print("Please select one of the following options")
        print("1. Add a new task")
        print("2. Delete a task")
        print("3. List tasks")
        print("4. Quit")

        choice = input("Enter your choice: ")

        if choice == "1":
            addTask()
            print()
        elif choice == "2":
            deleteTask()
        elif choice == "3":
            listTasks()
        elif choice == "4":
            break
        else:
            print("Wrong input.")

    print("close")
```
